```python
def create_and_update_fpo():
    start = time.time()
    dirName = "K:\\fpo\\"
    dirName2 = "K:\\fpo\\"
    if not os.path.exists(dirName2):
        os.makedirs(dirName2)
    chromeOptions = webdriver.ChromeOptions()
    prefs = {"download.default_directory" : dirName}
    chromeOptions.add_experimental_option("prefs",prefs)
    chromedriver = "./chromedriver"
    driver = webdriver.Chrome(executable_path=chromedriver, chrome_options=chromeOptions)
    driver.get("http://www.fpo.go.th/main/Statistic-Database.aspx")    
    # Navigate through the page
    # Data-Stat-list-wrapper


# lists for id charts


    all_class = driver.find_elements_by_class_name('DBStatisticExcelCsv')
    for i in all_class:
        try:
            href = i.find_element_by_tag_name('a')
            string = href.get_attribute('href')
#             print(string)
            if 'getattachment' in string:
    #             print('yes')
                href.click()
        except:
            pass
    driver.close()
    end = time.time()
    print('time execute: ', end - start)

    
        
```

```python
dirName = "K:\\fpo\\"
def remove_rename_dup():
    list_of_update = []
    for filename in os.listdir("K:\\fpo\\"):
        if '(' in filename:
            name = filename[:-8] + '.xls'
            list_of_update.append(name)

# Find if there is a file that contains those name -- then delete the file
    for filename in os.listdir("K:\\fpo\\"):
        for file in list_of_update:
            if filename == file:
                try:
                    os.remove("K:\\fpo\\" + filename)
                except:
                    pass

    for filename in os.listdir("K:\\fpo\\"):
        if '(' in filename:
            name = filename[:-8] + '.xls'
            src = "K:\\fpo\\" + filename
            name = "K:\\fpo\\" + name
            try:
                os.rename(src, name)
            except:
                os.remove(name)
                os.rename(src, name)
                

def change_eng_all():
    for filename in os.listdir("K:\\fpo\\"):
        if '(' in filename:
            os.remove("K:\\fpo\\" + filename)


    for filename in os.listdir("K:\\fpo\\"):
        if 'ENG_ALL' in filename:
            name = filename[:-13] + '.XLSX'
            src = "K:\\fpo\\" + filename
            name = "K:\\fpo\\" + name
            os.rename(src, name)
```
